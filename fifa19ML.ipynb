```python
#importing modules

import pandas as pd
```

```python
#opening the dataset and reading it

df = pd.read_csv('data.csv')
df
```

### <== Changing head here changes the graphs and data shown with the number given ==>

```python
#splitting the dataset and acquiring the data of the players

df1  = pd.DataFrame(df , columns=['Name','Age','Club','Nationality','Wage','Value']).head(20)
df1
```

```python
#converting the K,M,B as thousand,million,and billion respectively

def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
wage = df1['Wage'].replace('[€]','',regex = True).apply(value_to_float)
value = df1['Value'].replace('[€]','',regex = True).apply(value_to_float)


#changing the wage and value of player values with actual amount

df1['Wage']= wage
df1['Value']= value

#creating a Difference block with the difference between wage and value of player

df1['Difference'] = df1['Value']-df1['Wage'] 
df2 = df1.sort_values('Difference' , ascending=False)
df2[df2['Age']<35]
```

```python
#if we want to check for more than 20 players change the above head value as required,you can see the heading above /\
#                                                                                                                   |  

#Importing seaborn module for plotting purpose
#plotting a graph with wage and value such that we can see the credibility in the player with the salary he takes in the graph

import seaborn as sns
sns.set()
graph =sns.scatterplot(x = 'Wage',y = 'Value',data = df2)
```

```python
#importing modules required for hovering effect such that we can see the players names in the graph

import numpy as np
from bokeh.plotting import figure,show
from bokeh.models import HoverTool
from bokeh.io import output_notebook

output_notebook()
```

```python
#hovering effect was given

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage,Value)", "(@Wage,@Value)"),
    ("Name", "@Name"),("Club", "@Club")])

forhover = figure(title="fifa 2019", x_axis_label='Wage', y_axis_label='Value',tools =[TOOLTIPS])
forhover.circle('Wage','Value',size=10,source=df2)
show(forhover)

#click the hover image below the bokeh symbol to get a + cursor that hovers the players details 
```
